```python
cvd_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True, axis=1)
```

```python
cvd_df.head(10)
```

```python
cvd_df["Date"]=pd.to_datetime(cvd_df['Date'],format='%Y-%m-%d')
```

```python
cvd_df.head()
```

```python
#Active Cases
cvd_df['Active_Cases']=cvd_df['Confirmed']-(cvd_df['Deaths']+cvd_df['Cured'])
cvd_df.tail()
```

```python
statewise=pd.pivot_table(cvd_df,values = ["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise["Recovery_Rate"]=statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality_Rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#STATES WITH MOST ACTIVE CASES
top_active_cases=cvd_df.groupby(by = "State/UnionTerritory").max()[['Active_Cases','Date']].sort_values(by = ['Active_Cases'], ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(15,10))
```

```python
plt.title("Top 10 states with most active cases",size=25)
```

```python
ax=sns.barplot(data=top_active_cases.iloc[:10], y='Active_Cases',x="State/UnionTerritory", linewidth=2,edgecolor="red")
```

```python
#STATES WITH MOST ACTIVE CASES
top_active_cases=cvd_df.groupby(by = "State/UnionTerritory").max()[['Active_Cases','Date']].sort_values(by = ['Active_Cases'], ascending=False).reset_index()
fig=plt.figure(figsize=(15,10))
plt.title("Top 10 states with most active cases",size=25)
ax=sns.barplot(data=top_active_cases.iloc[:10], y='Active_Cases',x="State/UnionTerritory", linewidth=2,edgecolor="black")
plt.xlabel("States")
plt.ylabel("Active_Cases")
plt.show()
```

```python
#STATES WITH MOST DEATHS
top_deaths=cvd_df.groupby(by = "State/UnionTerritory").max()[['Deaths','Date']].sort_values(by = ['Deaths'], ascending=False).reset_index()
fig=plt.figure(figsize=(15,10))
plt.title("Top 10 states with most deaths",size=25)
ax=sns.barplot(data=top_deaths.iloc[:10], y='Deaths',x="State/UnionTerritory", linewidth=2,edgecolor="black")
plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```
